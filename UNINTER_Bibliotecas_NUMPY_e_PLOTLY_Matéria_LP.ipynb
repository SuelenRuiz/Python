# NUMPY - 7 Ordenação e Adição de elementos

```python
import numpy as np

rng=np.random.default_rng(seed=42)

array_1= rng.random([2,3])

print(f'Array 1 SEM ordenação: \n{array_1}\n')
print(f'Array 1 COM ordenação - ordenado por linha: \n{np.sort(array_1)}\n')
print(f'Array 1 COM ordenação - ordenado por coluna: \n{np.sort(array_1, axis=0)}\n')


# Contatenação

array_2=rng.random([2,1])
print(f'Array2:\n{array_2}\n')

array_concatenado = np.concatenate([array_1, array_2], axis=1)
print(f'Array concatenado:\n{array_concatenado}\n')

array_concatenado_2=np.hstack([array_1,array_2])
print(f'Array concatenado com hstack:\n{array_concatenado_2}\n')
```

# NUMPY - 9 Como mudar um formato de array

```python
import numpy as np

array_a= np.arange(0,20)
print(f'Array a: \n{a}\n')

# Duas formas diferentes de mudar o formato o array:

array_a = array_a.reshape(4,5)
print(f'Array a com formato alterado: \n{array_a}\n')

array_a = np.reshape(array_a,[4,5])
print(f'Array a com formato alterado: \n{array_a}\n')

# Para achar a matriz:

array_a = array_a.flatten()
print(f'Array a com formato alterado: \n{array_a}\n')

array_b = array_a.reshape(4,5)
print(f'Array b: \n{array_b}\n')

array_b = np.vsplit(array_b, 4)
print(f'Array b com formato alterado: \n{array_b}\n')

```

# NUMPY - 11 Salvando um array num arquivo txt

```python
import numpy as np

x = np.arange(27)
x = x.reshape((3,3,3))
print(f'tensor 3x3x3:\n {x}')

y = x.flatten()
print(f'\n vetor de 27 posições:\n {y}')
np.savetxt('arrayTeste.txt', y,fmt= '%d', delimiter = ';')

z= np.loadtxt('arrayTeste.txt',dtype = np.int8 ,delimiter = ';',)

print(f'\n arquivo recuperado e carregado para dentro do programa:\n {z} \ncom o seguinte tipo de dado: {(z.dtype)}')
```

# PLOTLY - 2 Criando arrays para montar

```python
import numpy as np

a = 3/4
b = 2/4
c = 1/4
vetor_a=np.ones(20)
vetor_b=np.ones(20)
vetor_c=np.ones(20)

for i in range(1,20):
  vetor_a[i]=vetor_a[i-1]+a**i
  vetor_b[i]=vetor_b[i-1]+b**i
  vetor_c[i]=vetor_c[i-1]+c**i
print(vetor_a)
print(vetor_b)
print(vetor_c)

np.savetxt('vetor_a.txt', vetor_a, fmt='%f', delimiter=';')
np.savetxt('vetor_b.txt', vetor_b, fmt='%f', delimiter=';')
np.savetxt('vetor_c.txt', vetor_c, fmt='%f', delimiter=';')
```

# PLOTLY - 3 Criando nosso primeiro gráfico

```python
import numpy as np
import plotly.express as px

array_a = np.loadtxt('vetor_a.txt', dtype=np.float64, delimiter=';')
array_b = np.loadtxt('vetor_b.txt', dtype=np.float64, delimiter=';')
array_c = np.loadtxt('vetor_c.txt', dtype=np.float64, delimiter=';')

print(array_a)


array_abc = np.vstack([array_a, array_b, array_c])
print(array_abc)
array_abc = array_abc.transpose()
print(array_abc)
fig=px.scatter(array_abc)
fig.show()
```

# PLOTLY - 5 Utilizando plotly.graph_objects

```python
import numpy as np
import plotly.graph_objects as go
array_a_axis_y = np.loadtxt("vetor_a.txt",dtype=np.float64, delimiter= ';')
array_b_axis_y = np.loadtxt("vetor_b.txt",dtype=np.float64, delimiter= ';')
array_c_axis_y = np.loadtxt("vetor_c.txt",dtype=np.float64, delimiter= ';')

array_axis_x = np.arange(1,array_a_axis_y.size)
fig = go.Figure()
fig.add_trace(go.Scatter(x= array_axis_x, y= array_a_axis_y,mode='lines', name='Série de 3/4'))
fig.add_trace(go.Scatter(x= array_axis_x, y= array_b_axis_y,mode='markers', name='Série de 2/4'))
fig.add_trace(go.Scatter(x= array_axis_x, y= array_c_axis_y,mode='markers+lines', name='Série de 1/4'))
fig.update_layout(title = "Somatório de Series com frações",
                  xaxis_title = "termos",
                  yaxis_title = "valor")
fig.show()
```
