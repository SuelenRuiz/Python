# Programação Orientada a Objetos (POO)

```python
class Bolo:
    peso=0

    def atribuir_peso(self, peso_a_atribuir):
        self.peso = peso_a_atribuir

bolo_aniversario=Bolo()
print(bolo_aniversario.peso)

bolo_aniversario.atribuir_peso(2)
print(bolo_aniversario.peso)
```

```python
bolo_casamento=Bolo()
bolo_casamento.atribuir_peso(5)
print(bolo_casamento.peso)
```

```python
class Bolo:
    peso=0

    def __init__(self,peso_inicial=0):
        self.peso = peso_inicial
    def atribuir_peso(self, peso_a_atribuir):
        self.peso = peso_a_atribuir

bolo_aniversario=Bolo()
print(bolo_aniversario.peso)

bolo_aniversario.atribuir_peso(2)
print(bolo_aniversario.peso)
```

```python
bolo_casamento=Bolo()
bolo_casamento.atribuir_peso(5)
print(bolo_casamento.peso)
```

# Classe, atributos, método construtor, outros métodos

```python
class Pessoa:
    # Atributos
    def __init__(self, nome, idade): # criação do método construtor __init__, que é chamado automaticamente quando uma nova instância da classe é criada. Ele recebe os parâmetros nome e idade, e os atribui aos atributos self.nome e self.idade da instância.
        self.nome = nome
        self.idade = idade
    # Método que usa os atributos para apresentar a pessoa
    def apresentar(self):
        print(f"Olá, meu nome é {self.nome} e tenho {self.idade} anos.")
```

```python
# Duas instâncias da classe Pessoa
Suelen = Pessoa("Suelen", 32)
Suelen.apresentar()

Felipe = Pessoa("Felipe", 30)
Felipe.apresentar()
```

# Herança
Herança: uma classe pode herdar atributos e métodos de outra classe. A classe que herda é chamada de classe filha ou subclasse, e a classe que é herdada é chamada de classe base ou superclasse.


```python
# A classe mãe é a classe Pessoa, e a classe filha é a classe Funcionario. A classe Funcionario herda os atributos nome e idade, e o método apresentar da classe Pessoa. Além disso, a classe Funcionario tem um atributo adicional chamado salario, e um método apresentar que chama o método apresentar da classe Pessoa e depois imprime uma mensagem adicional.
class Funcionario(Pessoa):
    pass # A palavra-chave pass é usada para indicar que a classe Funcionario não tem nenhum código adicional além do que é herdado da classe Pessoa.
    # O python entende que não terá definição e segue sem dar erro.

f1= Funcionario("Ricardo", 18)
f1.apresentar() # O método apresentar é herdado da classe Pessoa, então ele pode ser chamado para a instância f1 da classe Funcionario.  
```

```python
class Funcionario(Pessoa):
    def __init__(self, nome, idade, salario):
        super().__init__(nome, idade) # A função super() é usada para chamar o método __init__ da classe base (Pessoa) e inicializar os atributos nome e idade.
        self.salario = salario # O atributo salario é inicializado diretamente na classe Funcionario.

    def apresentar(self):
        super().apresentar() # Chama o método apresentar da classe base (Pessoa) para imprimir o nome e a idade.
        print(f"Eu sou um funcionário e meu salário é {self.salario}.") # Imprime uma mensagem adicional com o salário.
```

```python
f1= Funcionario("Ricardo", 18, 1200)
f1.apresentar()
```

```python
# Temos dois métodos "apresentar". Qual método o programa escolhe?
## O Python sempre escolhe o método mais específico, ou seja, escolherá o referente à classe "Funcionario".

# O método está sendo sobrescrito. Ou seja, queremos que o método tenha um comportamento diferente do método base.

class Funcionario(Pessoa):
    #pass
    def __init__(self,nome,idade,cadastro):
        self.nome= nome
        self.idade= idade
        self.cadastro= cadastro
    def apresentar(self):
        print(f"Olá, sou o funcionário(a) {self.nome} e minha idade é {self.idade} anos.")

f1=Funcionario("Ricardo", 18, "12345")
f2=Funcionario("Maria", 25, "54321")

f1.apresentar()
f2.apresentar()
print(f1.cadastro) # Acessando atributo específico do Ricardo.
print(f2.cadastro) # Acessando atributo específico da Maria.
```

```python
class Cliente(Pessoa):
    # Variável global estática para a classe "Cliente".
    quantidadeClientes = 0
    def __init__(self, nome, idade):
        super().__init__(nome, idade) # Referenciando a classe base/mãe "Pessoa".
        Cliente.quantidadeClientes+=1
        self.cadastro=1000+Cliente.quantidadeClientes # O cadastro é único para cada cliente, portanto, deve-se usar self. e não Cliente.
         # O cadastro é gerado automaticamente com base na quantidade de clientes já criados.

    def apresentar(self):
        super().apresentar() # Chama o método apresentar da classe mãe "Pessoa".
        print(f"Eu sou cliente de cadastro {self.cadastro}. E a quantidade de clientes é {Cliente.quantidadeClientes}.") # Imprime o cadastro do cliente e a quantidade total de clientes.

c1= Cliente("Ana", 28)
c2= Cliente("Bruno", 35)
c3= Cliente("Carla", 22)
c1.apresentar()
c2.apresentar()
c3.apresentar()
```

Ao final, apresentar na tela os alunos e a média por turma.




```python
class Aluno:
    totalAlunos = 0 # serve para contar quantos alunos já foram criados
    def __init__(self, nome, nota):
        self.nome = nome
        self.nota = nota #0-100
        Aluno.totalAlunos += 1 # Incrementa a quantidade total de alunos sempre que um novo aluno é criado.
        self.ru=1000+Aluno.totalAlunos # O número de registro do aluno é gerado automaticamente com base na quantidade total de alunos.
    def info(self):
        print(f'Nome: {self.nome}, Nota: {self.nota}, RU: {self.ru}, Total de Alunos: {Aluno.totalAlunos}')
    def getNota (self):
        return self.nota


a1= Aluno("João", 85)
a2= Aluno("Maria", 92)
a3= Aluno("Pedro", 78)
# a1.info()
# a2.info()
# a3.info()
   
class Turma:
   #pass
   def __init__(self,nome,limiteAlunos):
       self.nome= nome
       self.limiteAlunos= limiteAlunos
       self.listaAlunos= [] # Lista para armazenar os alunos matriculados na turma.

   def addAluno(self, aluno):
       if len(self.listaAlunos)<self.limiteAlunos:
           self.listaAlunos.append(aluno) # Adiciona o aluno à lista de alunos da turma.
           return True
       return False # Retorna False se a turma já atingiu o limite de alunos.
   
   def mediaTurma(self):
      soma=0
      for aluno in self.listaAlunos:
         soma+= aluno.getNota() # Acessa a nota de cada aluno usando o método getNota() e acumula a soma das notas.
         return soma/len(self.listaAlunos) # Retorna a média das notas da turma, dividindo a soma das notas pelo número de alunos na turma.
    
   def listarAlunos(self):
      print(f'A turma {self.nome.upper()} possui os seguintes alunos: ')
      for aluno in self.listaAlunos:
         print(f'Nome: {aluno.nome}, Nota: {aluno.nota}') # Imprime o nome e a nota de cada aluno na turma.
      print(f'-----Média da turma: {self.mediaTurma():.2f} \n\n ') # Imprime a média das notas da turma usando o método mediaTurma().


turma1 = Turma('Programação', 5)
turma2 = Turma('Matemática', 3)
turma1.addAluno(a1)
turma1.addAluno(a2)
turma1.addAluno(a3)
turma2.addAluno(a1)
turma2.addAluno(a3)

# print(turma1.addAluno(a3)) # A turma já atingiu o limite de alunos, então este aluno não será adicionado.
# print(turma1.mediaTurma()) # Imprime a média das notas dos alunos na turma
turma1.listarAlunos() # Imprime o nome e a nota de cada aluno na turma
# turma1.mediaTurma()

turma2.listarAlunos()
# turma2.mediaTurma()
# print(turma1.listaAlunos[0].nome)
```
